```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
sns.set()
```

```python
np.random.seed(123)
center_X = [0.2, 0.5, 0.8]
center_y = [0.8, 0.4, 0.9]
X = np.random.randn(300)*0.125 + np.resize(center_X, 300)
y = np.random.randn(300)*0.1 + np.resize(center_y, 300)
data = np.array([X, y]).T
```

```python
kmeans = KMeans(n_clusters = 3,
                n_jobs = -1,
                random_state = 123).fit(data)
```

```python
df = pd.DataFrame(data, columns = ["x", "y"])
df["label"] = kmeans.labels_
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,5))

title = ["Raw Data", "Clustered Data"]

idx, ax = 0, axes[0]
sns.scatterplot(X, y, ax = ax)
ax.set_title(title[idx].upper(), fontsize = 20, fontweight = "bold")
ax.set_xticks([])
ax.set_yticks([])

idx, ax = 1, axes[1]
sns.scatterplot(data = df, x = "x", y = "y", hue = "label", ax = ax, palette="plasma")
ax.set_title(title[idx].upper(), fontsize = 20, fontweight = "bold")
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel("")
ax.set_ylabel("")
ax.get_legend().remove()
```
